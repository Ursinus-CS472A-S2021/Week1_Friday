## Week 1: Digital Audio in Numpy
We started off by doing some simple examples of digital audio in numpy

```python
import numpy as np # This is the main numerical library we will use
import matplotlib.pyplot as plt # This is the main plotting library we will use
#import librosa # We will use this library to load in audio
import IPython.display as ipd # This is a library that allows us to play audio samples in Jupyter
%matplotlib inline
```

```python
sr = 44100 # Audio sample rate
x = np.arange(sr)
# We can make a "sawtooth wave" by taking the 
# element-wise mod of arange by some number
y = x % 100  # remainder of every elem after /100
plt.plot(y[0:400])
ipd.Audio(y, rate=sr)
## Takehome message
## The faster you repeat your pattern, the higher the pitch
```

```python
2*2*3*3*5*5*7*7 ## 44100 has an interesting prime decomposition!
## We want something that's at least 40,000, and this fits the bill
```

```python
plt.stem(np.random.rand(50))  # We often think of digital audio
# as samples/measurements taken at a particular point in time,
# so stem plots represent that well
```

```python
# Random numbers in an array sound like static/noise!
# Usually this is a nuisance, but we can use it as a source
# to make sounds later
y = np.random.rand(sr)
ipd.Audio(y, rate=sr)
```

```python
# Here's a linear fade of a sawtooth wave (Manning's idea)

sr = 44100 # Audio sample rate
x = np.arange(sr)
y = x % 100 
y = (x % 100)*(sr-1-x) # Fade out with an element-wise multiplication
plt.plot(y)
ipd.Audio(y, rate=sr)
```

```python
# Here's a "hyperbolic" or 1/x fade (Tom's idea)
sr = 44100 # Audio sample rate
x = np.arange(sr*2) 
y = (x % 100)/(0.1 + x/sr) # Fade out with an element-wise multiplication
plt.plot(y)
ipd.Audio(y, rate=sr)
```
